```python
import pandas as pd
from pandas import DataFrame


PATH = '../data/dev.csv'
df = pd.read_csv(PATH)
df.head(15)
```

```python
slack = pd.concat([df[df.source == ('slack-sampled')], df[df.source == ('slack-rtt')]])
petition = pd.concat([df[df.source == ('petition-sampled')], df[df.source == ('petition-rtt')]])
nsmc = pd.concat([df[df.source == ('nsmc-sampled')], df[df.source == ('nsmc-rtt')]])
slack.to_csv('../data/dev_slack.csv')
petition.to_csv('../data/dev_petition.csv')
nsmc.to_csv('../data/dev_nsmc.csv')
```
